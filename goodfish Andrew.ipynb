```python
from thunder import PCA
modelPCA = PCA(k=5,svdmethod='em').fit(data)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')
sns.set_style('darkgrid')
plt.plot(modelPCA.comps.T);
```

```python
from thunder import KMeans
modelK = KMeans(k=30).fit(data)
from thunder import Colorize
sns.set_style('darkgrid')
newclrs = Colorize.optimize(modelK.centers, ascmap=True)
plt.gca().set_color_cycle(newclrs.colors)
plt.plot(modelK.centers.T);
```

```python
plt.plot(data.normalize().max());
plt.plot(data.normalize().mean());
plt.plot(data.normalize().min());
```

```python
pts = modelPCA.scores.subset(2000, thresh=0.01, stat='norm')
sns.set_style('darkgrid')
clrs = Colorize(totype='polar', scale=50).points(pts[:,0:2])
plt.scatter(pts[:,0],pts[:,1], c=clrs, s=75, alpha=0.7);
```

```python
from numpy import asarray
recon = asarray(map(lambda x: (x[0] * modelPCA.comps[0, :] + x[1] * modelPCA.comps[1, :]).tolist(), pts))
plt.gca().set_color_cycle(clrs)
plt.plot(recon.T);
```
